```python
#import dependencies and setup
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#import and read clinical trial and mouse drug csv files
clinical_data_df = pd.read_csv('data/clinicaltrial_data.csv')
mouse_data_df = pd.read_csv('data/mouse_drug_data.csv')
```

```python
#identify duplicates 
duplicate_id = clinical_data_df.loc[clinical_data_df.duplicated(subset=['Mouse ID','Timepoint']),'Mouse ID'].unique()
```

```python
#drop duplicates before merging
clean_clinical_data_df = clinical_data_df[clinical_data_df['Mouse ID'].isin(duplicate_id)==False]
clean_mouse_data_df = mouse_data_df[mouse_data_df['Mouse ID'].isin(duplicate_id)==False]
```

```python
#combine and display data into single dataset
combined_data_df = pd.merge(clean_clinical_data_df, clean_mouse_data_df,how = 'outer',on='Mouse ID')
combined_data_df.head()
```

```python
#list(combined_data_df.columns)
```

```python
#group combined data by Drug and Timepoint
combined_data_df_grouped = combined_data_df.groupby(['Drug','Timepoint'])
```

```python
#store the mean of grouped tumor volumes 
#convert grouped data into dataframe and display
tumor_volume_mean_df = pd.DataFrame(combined_data_df_grouped['Tumor Volume (mm3)'].mean())
tumor_volume_mean_df.head()
```

```python
#store the standard error of grouped tumor volume
#convert grouped data into dataframe and display
tumor_volume_standard_error_df = pd.DataFrame(combined_data_df_grouped['Tumor Volume (mm3)'].sem())
tumor_volume_standard_error_df.head()
```

```python
#minor data munging to re-format the dataframes
tumor_volume_mean_df_formatted = tumor_volume_mean_df.reset_index()
tumor_volume_mean_pivot = tumor_volume_mean_df_formatted.pivot(index='Timepoint',columns='Drug')['Tumor Volume (mm3)']
tumor_volume_mean_pivot.head()
```

```python
#minor data munging to re-format the dataframe
tumor_volume_standard_error_df_formatted = tumor_volume_standard_error_df.reset_index()
tumor_volume_standard_error_pivot = tumor_volume_standard_error_df_formatted.pivot(index='Timepoint',columns='Drug')['Tumor Volume (mm3)']
tumor_volume_standard_error_pivot.head()
```

```python
# generate plot with error bars
plt.figure(figsize=(14,8))
x_axis = np.arange(0,50,10)
plt.xlim = 45


plt.errorbar(tumor_volume_mean_pivot.index, tumor_volume_mean_pivot["Capomulin"],yerr=tumor_volume_standard_error_pivot['Capomulin'],\
             fmt='o--')
plt.errorbar(tumor_volume_mean_pivot.index, tumor_volume_mean_pivot["Ceftamin"],yerr=tumor_volume_standard_error_pivot['Ceftamin'],\
             fmt='+--')
plt.errorbar(tumor_volume_mean_pivot.index, tumor_volume_mean_pivot["Infubinol"],yerr=tumor_volume_standard_error_pivot['Infubinol'],\
             fmt='^--')
plt.errorbar(tumor_volume_mean_pivot.index, tumor_volume_mean_pivot["Ketapril"],yerr=tumor_volume_standard_error_pivot['Ketapril'],\
             fmt='s--')
plt.errorbar(tumor_volume_mean_pivot.index, tumor_volume_mean_pivot["Naftisol"],yerr=tumor_volume_standard_error_pivot['Naftisol'],\
             fmt='p--')
plt.errorbar(tumor_volume_mean_pivot.index, tumor_volume_mean_pivot["Placebo"],yerr=tumor_volume_standard_error_pivot['Placebo'],\
             fmt='d--')
plt.errorbar(tumor_volume_mean_pivot.index, tumor_volume_mean_pivot["Propriva"],yerr=tumor_volume_standard_error_pivot['Propriva'],\
             fmt='x--')
plt.errorbar(tumor_volume_mean_pivot.index, tumor_volume_mean_pivot["Ramicane"],yerr=tumor_volume_standard_error_pivot['Ramicane'],\
             fmt='|--')
plt.errorbar(tumor_volume_mean_pivot.index, tumor_volume_mean_pivot["Stelasyn"],yerr=tumor_volume_standard_error_pivot['Stelasyn'],\
             fmt='_--')
plt.errorbar(tumor_volume_mean_pivot.index, tumor_volume_mean_pivot["Zoniferol"],yerr=tumor_volume_standard_error_pivot['Zoniferol'],\
             fmt='*--')

#set title and axis labels 
plt.title('Tumor Response to Treatment',fontsize=20)
plt.xlabel('Time (days)',fontsize=14)
plt.ylabel('Tumor Volume(mm3)',fontsize=14)

#add legend
plt.legend(loc='best')
plt.grid()

#save the figure
plt.savefig("Images/tumorResponsetoTreatment.png")

#show plot
plt.show()
```

```python
# generate plot with error bars
plt.figure(figsize=(14,8))
x_axis = np.arange(0,50,10)
plt.xlim = 45

plt.errorbar(metastatic_response_mean_df_pivot.index, metastatic_response_mean_df_pivot["Capomulin"],yerr=metastatic_response_standard_error_df_pivot['Capomulin'],\
             marker='o',linestyle='dashed')
plt.errorbar(metastatic_response_mean_df_pivot.index, metastatic_response_mean_df_pivot["Ceftamin"],yerr=metastatic_response_standard_error_df_pivot['Ceftamin'],\
             marker='+',linestyle='dashed')
plt.errorbar(metastatic_response_mean_df_pivot.index, metastatic_response_mean_df_pivot["Infubinol"],yerr=metastatic_response_standard_error_df_pivot['Infubinol'],\
             marker='^',linestyle='dashed')
plt.errorbar(metastatic_response_mean_df_pivot.index, metastatic_response_mean_df_pivot["Ketapril"],yerr=metastatic_response_standard_error_df_pivot['Ketapril'],\
             marker='s',linestyle='dashed')
plt.errorbar(metastatic_response_mean_df_pivot.index, metastatic_response_mean_df_pivot["Naftisol"],yerr=metastatic_response_standard_error_df_pivot['Naftisol'],\
             marker='p',linestyle='dashed')
plt.errorbar(metastatic_response_mean_df_pivot.index, metastatic_response_mean_df_pivot["Placebo"],yerr=metastatic_response_standard_error_df_pivot['Placebo'],\
             marker='d',linestyle='dashed')
plt.errorbar(metastatic_response_mean_df_pivot.index, metastatic_response_mean_df_pivot["Propriva"],yerr=metastatic_response_standard_error_df_pivot['Propriva'],\
             marker='x',linestyle='dashed')
plt.errorbar(metastatic_response_mean_df_pivot.index, metastatic_response_mean_df_pivot["Ramicane"],yerr=metastatic_response_standard_error_df_pivot['Ramicane'],\
             marker='|',linestyle='dashed')
plt.errorbar(metastatic_response_mean_df_pivot.index, metastatic_response_mean_df_pivot["Stelasyn"],yerr=metastatic_response_standard_error_df_pivot['Stelasyn'],\
             marker='_',linestyle='dashed')
plt.errorbar(metastatic_response_mean_df_pivot.index, metastatic_response_mean_df_pivot["Zoniferol"],yerr=metastatic_response_standard_error_df_pivot['Zoniferol'],\
             marker='*',linestyle='dashed')


#set title and axis labels 
plt.title('Metastatic Spread During Treatment',fontsize=20)
plt.xlabel('Treatment Duration (Days)',fontsize=14)
plt.ylabel('Metastatic Sites',fontsize=14)

#add legend
plt.legend(loc='upper left')
plt.grid()

#save figure
plt.savefig("Images/metastaticSpreadDuringTreatment.png")

#show plot
plt.show()
```

```python
#calculate the percent changes for each drug
percent_change = ((tumor_volume_mean_pivot.iloc[-1]-tumor_volume_mean_pivot.iloc[0])/tumor_volume_mean_pivot.iloc[0]) * 100

#display the data to confirm
percent_change
```

```python
#store all relevant percent changes into a tuple
changes = (percent_change['Capomulin'],
           percent_change['Ceftamin'],
           percent_change['Infubinol'],
           percent_change['Ketapril'],
           percent_change['Naftisol'],
           percent_change['Placebo'],
           percent_change['Propriva'],
           percent_change['Ramicane'],
           percent_change['Stelasyn'],
           percent_change['Zoniferol'],
           )
```

```python
plt.figure(figsize=(16,8))
x_axis_labels = percent_change.index
plt.xticks(np.arange(len(x_axis_labels)), x_axis_labels, fontsize=14)

plt.bar(np.arange(10),
       changes,
       color = ['r' if changes[r] > 0 else 'g' for r in np.arange(len(x_axis_labels))])

#
xcount = 0
for r in changes:
    if r < 0:
        y_coor = -3.5
    else:
        y_coor = 2
    plt.text(xcount,y_coor, str(round(r,1)) + '%', ha = 'center', color='white')
    xcount +=1


plt.title('Tumor Change Over 45 Day Treatment', fontsize=20)
plt.ylabel('% Tumor Volume change', fontsize=14)
plt.axhline(y=0, color='black')


#save the figure
plt.savefig('Images/tumorChangeOver45DayTreatment.png')


#show the figure
plt.show()
```
